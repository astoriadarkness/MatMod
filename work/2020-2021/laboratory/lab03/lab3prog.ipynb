```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math
```

# по условию из варианта 70

```python
x0 = 80000
y0 = 60000
t0 = 0
tmax = 1

a = 0.21
b = 0.855
c = 0.455
h = 0.32

a2 = 0.267
b2 = 0.687
c2 = 0.349
h2 = 0.491
```

```python
def P(t):
    p = math.sin(t)+2
    return p

def Q(t):
    q = math.cos(t)+2
    return q
```

```python
def P2(t):
    p = math.fabs(math.sin(2*t))
    return p

def Q2(t):
    q = 2*math.fabs(math.cos(t))
    return q

def f(y, t):
    y1, y2 = y
    return [-a*y1 - b*y2 + P(t), -c*y1 - h*y2 + Q(t) ]

def f2(y, t):
    y1, y2 = y
    return [-a2*y1 - b2*y2 + P2(t), -c2*y1*y2 - h2*y2 + Q2(t) ]
```

# строим графики

```python
t = np.linspace( 0, tmax, num = 100)
y0 = [x0, y0]
w1 = odeint(f, y0, t)
y11 = w1[:,0]
y21 = w1[:,1]
fig = plt.figure(facecolor='white')
plt.plot(t, y11, t, y21, linewidth=2)
plt.ylabel("x, y")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig.savefig('1.png', dpi = 800)
```

```python
w1 = odeint(f2, y0, t)
y12 = w1[:,0]
y22 = w1[:,1]
fig2 = plt.figure(facecolor='white')
plt.plot(t, y12, t, y22, linewidth=2)
plt.ylabel("x, y")
plt.xlabel("t")
plt.grid(True)
plt.show()
fig2.savefig('2.png', dpi = 800)
```
