```python
import numpy as np
from scipy. integrate import odeint
import matplotlib.pyplot as plt
import math


N = 14041 #общая численность популяции
I0 = 131 #количество инфицированных в начальный момент времени
R0 = 71 #количество здоровых с иммуенитетом в начальный момент времени
S0 = N-I0-R0 #количество восприимчивых к болезни в начальный момент времени

a = 0.01 #коэффициент заболеваемости
b = 0.02 #коэфициент выздоровления


x0 = [S0, I0, R0]

#функция для первого случая
def syst(y, t):
    y1, y2, y3 = y
    return [0, -b*y2, b*y2 ]

#функция для второго случая
def syst2(y, t):
    y1, y2, y3 = y
    return [-a*y1, a*y1-b*y2, b*y2 ]

t = np.arange( 0, 200, 0.01)
y1 = odeint(syst, x0, t)
y1s = y1[:,0]
y1i = y1[:,1]
y1r = y1[:,2]

#для первого случая
fig = plt.figure(facecolor='white')
plt.plot(t, y1s, linewidth=2, label='S(t)')
plt.plot(t, y1i, linewidth=2, label='I(t)')
plt.plot(t, y1r, linewidth=2, label='R(t)')
plt.ylabel("численность людей")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
fig.savefig('1.png', dpi = 600)

#для второго случая
y2 = odeint(syst2, x0, t)
y2s = y2[:,0]
y2i = y2[:,1]
y2r = y2[:,2]

fig2 = plt.figure(facecolor='white')
plt.plot(t, y2s, linewidth=2, label='S(t)')
plt.plot(t, y2i, linewidth=2, label='I(t)')
plt.plot(t, y2r, linewidth=2, label='R(t)')
plt.ylabel("численность людей")
plt.xlabel("t")
plt.grid(True)
plt.legend()
plt.show()
fig2.savefig('2.png', dpi = 600)
```
